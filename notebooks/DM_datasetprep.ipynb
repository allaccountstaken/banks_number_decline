```python
#Load Net-Interest margin data for 4 peer groups
nim_under100M_df = pd.read_csv('../data/US100NIM.csv') #banks with assets under $100M 
nim_small_df = pd.read_csv('../data/US13NIM.csv') #banks with assets between $100-300M
nim_mid_df = pd.read_csv('../data/US31NIM.csv') #banks with assets between $300-1000M
nim_big_df = pd.read_csv('../data/US115NIM.csv') #banks with assets beteen $1-13B
nim_greater1500M_df = pd.read_csv('../data/USG15NIM.csv') #banks with assets greater than $1B
```

```python
nim_banks_df = nim_under100M_df.merge(nim_small_df, how = 'left')
nim_banks_df = nim_banks_df.merge(nim_mid_df, how = 'left')
nim_banks_df = nim_banks_df.merge(nim_big_df, how = 'left')
nim_banks_df = nim_banks_df.merge(nim_greater1500M_df, how = 'left')
nim_banks_df.head()
#banks_df
```

```python
#Load number of banks data for the 4 bank peer groups
num_under100M_df = pd.read_csv('../data/US100NUM.csv')
num_small_df = pd.read_csv('../data/US13NUM.csv')
num_mid_df = pd.read_csv('../data/US31NUM.csv')
num_big_df = pd.read_csv('../data/US115NUM.csv')
num_greater1500M_df = pd.read_csv('../data/USG15NUM.csv')
```

```python
num_banks_df = num_under100M_df.merge(num_small_df, how = 'left')
num_banks_df = num_banks_df.merge(num_mid_df, how = 'left')
num_banks_df = num_banks_df.merge(num_big_df, how = 'left')
num_banks_df = num_banks_df.merge(num_greater1500M_df, how = 'left')
num_banks_df.tail()
```

```python
#Load loan loss reserves banks data for the 4 bank peer groups
llr_under100M_df = pd.read_csv('../data/US100LLRTL.csv')
llr_small_df = pd.read_csv('../data/US13LLRTL.csv')
llr_mid_df = pd.read_csv('../data/US31LLRTL.csv')
llr_big_df = pd.read_csv('../data/US115LLRTL.csv')
llr_greater1500M_df = pd.read_csv('../data/USG15LLRTL.csv')
llr_banks_df = llr_under100M_df.merge(llr_small_df, how = 'left')
llr_banks_df = llr_banks_df.merge(llr_mid_df, how = 'left')
llr_banks_df = llr_banks_df.merge(llr_big_df, how = 'left')
llr_banks_df = llr_banks_df.merge(llr_greater1500M_df, how = 'left')
# llr_banks_df.tail()
banks_df = banks_df.merge(llr_banks_df, how = 'left')
banks_df.head()
```

```python
#Load return on assets banks data for the 4 bank peer groups
roa_under100M_df = pd.read_csv('../data/US100ROA.csv')
roa_small_df = pd.read_csv('../data/US13ROA.csv')
roa_mid_df = pd.read_csv('../data/US31ROA.csv')
roa_big_df = pd.read_csv('../data/US115ROA.csv')
roa_greater1500M_df = pd.read_csv('../data/USG15ROA.csv')
roa_banks_df = roa_under100M_df.merge(roa_small_df, how = 'left')
roa_banks_df = roa_banks_df.merge(roa_mid_df, how = 'left')
roa_banks_df = roa_banks_df.merge(roa_big_df, how = 'left')
roa_banks_df = roa_banks_df.merge(roa_greater1500M_df, how = 'left')
roa_banks_df.tail()
```

```python
#Load return on equity banks data for 4 bank peer groups
roe_under100M_df = pd.read_csv('../data/US100ROE.csv')
roe_small_df = pd.read_csv('../data/US13ROE.csv')
roe_mid_df = pd.read_csv('../data/US31ROE.csv')
roe_big_df = pd.read_csv('../data/US115ROE.csv')
roe_greater1500M_df = pd.read_csv('../data/USG15ROE.csv')
roe_banks_df = roe_under100M_df.merge(roe_small_df, how = 'left')
roe_banks_df = roe_banks_df.merge(roe_mid_df, how = 'left')
roe_banks_df = roe_banks_df.merge(roe_big_df, how = 'left')
roe_banks_df = roe_banks_df.merge(roe_greater1500M_df, how = 'left')
#roe_banks_df.tail()
banks_df = banks_df.merge(roe_banks_df, how = 'left')
banks_df.tail()
```

```python
#Load non-performing loans banks data for the 4 bank peer groups
npl_under100M_df = pd.read_csv('../data/US100NPTL.csv')
npl_small_df = pd.read_csv('../data/US13NPTL.csv')
npl_mid_df = pd.read_csv('../data/US31NPTL.csv')
npl_big_df = pd.read_csv('../data/US115NPTL.csv')
npl_greater1500M_df = pd.read_csv('../data/USG15NPTL.csv')
npl_banks_df = npl_under100M_df.merge(npl_small_df, how = 'left')
npl_banks_df = npl_banks_df.merge(npl_mid_df, how = 'left')
npl_banks_df = npl_banks_df.merge(npl_big_df, how = 'left')
npl_banks_df = npl_banks_df.merge(npl_greater1500M_df, how = 'left')
#npl_banks_df.tail()
banks_df = banks_df.merge(npl_banks_df, how = 'left')
banks_df.head()
```

```python
#Read and load aggregate data for all the banks
NI_df = pd.read_csv('../data/USNIM.csv')
ROA_df = pd.read_csv('../data/USROA.csv')
NP_df = pd.read_csv('../data/USNP.csv')
TAST_df = pd.read_csv('../data/USTAST.csv')

banks_df = banks_df.merge(NI_df, how = 'left')
banks_df = banks_df.merge(ROA_df, how = 'left')
banks_df = banks_df.merge(NP_df, how = 'left')
banks_df = banks_df.merge(TAST_df, how = 'left')
banks_df.shape
```
